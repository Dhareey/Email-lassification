```python
import numpy as np
import pandas as pd
import os
import numpy as np
from collections import Counter
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
# First, create a bag of words with all the words
```

```python
# The location of the dataset to use
directory = 'data1/train-mails'
```

```python
bag_of_words = []
files = [os.path.join(directory,f) for f in os.listdir(directory)]
for file in files:
    with open(file) as m:
        for line in m:
            sentence = word_tokenize(line)
            for word in sentence:
                if len(word) == 1:
                    pass
                elif word.isalpha() ==False:
                    pass
                elif word in stopwords.words():
                    pass
                else:
                    bag_of_words.append(word)
dictionary = Counter(bag_of_words)
```

```python
# Total length of the dictonary is 16,000. Use the first 6,000
dictionary = dictionary.most_common(6000)
```

```python
dictionary
```

```python
# Explore how each mail is
test =  files[0]
with open(test) as s:
    for i,line in enumerate(s):
        if i ==2:
            print(line)
```

```python
files[1]
```

```python
#Make a numpy array of all zeros with 6000 columns and the same number of rows in files
dataArray = np.zeros((len(files),len(dictionary)))
```

```python
dataArray
```

```python
# Create a numpy array of zeros for the target variables
emailTarget = np.zeros(len(files))
```

```python
emailTarget
```

```python
dictcheck =[row[0] for row in dictionary]
```

```python
dictcheck.index('doe')
```

```python
# Dealing with all the files now
rownumber = 0
for file in files:
    pred = file.replace('data1/train-mails\\','')
    if 'spmsg' in pred:
        emailTarget[rownumber] = 1
    with open(file) as eachtext:
        for i, line in enumerate(eachtext):
            if i == 2:
                filebag = line.split()
                for eachword in filebag:
                    if eachword in dictcheck:
                        colnumber = dictcheck.index(eachword)
                        dataArray[rownumber,colnumber] = filebag.count(eachword)
    
    rownumber+=1
```

```python
dataArray
```

```python
#Check how it is with pandas
dataframe = pd.DataFrame(dataArray, columns =dictcheck)
```

```python
dataframe
```

```python
dataframe.iloc[0]
```

```python
# 1 = Spam emails
emailTarget
```

```python
# Train naive bayes algorithm
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(dataArray, emailTarget)
```

```python
# Test the algorithm
testArray = np.zeros((len(files),len(dictionary)))
testpred = np.zeros(len(files))
test_dir = 'data1/test-mails'
files = [os.path.join(test_dir, c) for c in os.listdir(test_dir)]
rownumber = 0
for file in files:
    pred = file.replace('data1/test-mails\\','')
    if 'spmsg' in pred:
        testpred[rownumber] = 1
    with open(file) as eachtext:
        for i, line in enumerate(eachtext):
            if i == 2:
                filebag = word_tokenize(line)
                for eachword in filebag:
                    if eachword in dictcheck:
                        colnumber = dictcheck.index(eachword)
                        testArray[rownumber,colnumber] = filebag.count(eachword)
    
    rownumber+=1
```

```python
testArray
```

```python
predicted_labels = model.predict(testArray)
```

```python
predicted_labels
```

```python
#Evaluate the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(testpred, predicted_labels)
```
